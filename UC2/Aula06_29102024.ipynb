Aplicação de Resumos Estatísticos e Principais Consultas em um DataFrame Externo:

```python
import pandas as pd
import numpy as np
```

```python
dados=pd.read_csv('weight_change_dataset.csv')
display(dados.head())
estatistica=dados.describe()
display(estatistica)
df_copia=dados.copy() #BOAS PRÁTICAS: sempre inicie com uma cópia do dataset original para as primeiras manipulações/edições de dados
display(df_copia)
```

Consultas:

```python
#Selecionando todos os Níveis de Qualidade de Sono que sejam 'Excellent':
df_loc=df_copia.loc[df_copia['Sleep Quality']=='Excellent']
display(df_loc)
```

```python
#Selecionando a informação por índices (linhas e/ou colunas específicas):
df_loc2=df_copia.iloc[:2,:4]
display(df_loc2)
```

```python
#Selecionando A Partir de Filtros com Teste Lógico:
df_query=df_copia.query('Age<40')
display(df_query)
```
